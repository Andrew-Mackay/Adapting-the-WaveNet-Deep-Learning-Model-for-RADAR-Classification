```
import tensorflow as tf
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD
import numpy as np 

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
```

```
model = Sequential()
model.add(Dense(8, activation=tf.nn.relu, input_dim=2)) # Add layer with 8 neurons that use relu for activation function
model.add(Dense(8, activation=tf.nn.relu)) # Add layer with 8 neurons that use relu for activation function
model.add(Dense(1, activation='sigmoid')) # 10 for final layer (0-9) and use softmax as probability distribution
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```
history = model.fit(X, y, epochs=150, verbose=0) # verbose = 0 disables output
```

```
import matplotlib.pyplot as plt

print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```
# summarize history for loss
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

As the loss looks like it's still droppping, increase the number of epochs to see where it levels out

```
history = model.fit(X, y, epochs=1000, verbose=0) # verbose = 0 disables output
```

```
# summarize history for loss
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
